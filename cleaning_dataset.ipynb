```python
import pandas as pd
```

## Read the file

```python
df = pd.read_csv("data_with_spotify.csv") 
df
```

## Delete  unnecessary columns

```python
columns_to_delete = ['Unnamed: 0', 'direct_qualifier_10', 'favourite_10', 'host_10', 'key_change_10', 'instrument_10',
                      'qualified_10', "race", "selection", "age", 'final_draw_position', 'final_televote_points',
                     'final_jury_points',
                     'final_televote_votes', 'final_jury_votes', 'final_place',
                     'final_total_points', 'semi_televote_points',
                     'semi_jury_points', 'semi_total_points', 'semi_final', 'semi_draw_position']

# Delete the columns
df.drop(columns=columns_to_delete, inplace=True)
```

## Creating English language dummies

```python
df['eng_lan'] = df['language'].apply(lambda x: 1 if x == 'English' else 0)
df['inc_eng_lan'] = ((df['language'].str.contains('English')) & (df['eng_lan'] == 0)).astype(int)
```

## Dummies for styles and gender

```python
df['Male'] = df['gender'].apply(lambda x: 1 if x == 'Male' else 0)
df['Female'] = df['gender'].apply(lambda x: 1 if x == 'Female' else 0)

# add dummies for style

df['Pop'] = df['style'].apply(lambda x: 1 if x == 'Pop' else 0)
df['Rock'] = df['style'].apply(lambda x: 1 if x == 'Rock' else 0)
df['Ballad'] = df['style'].apply(lambda x: 1 if x == 'Ballad' else 0)
df['Traditional'] = df['style'].apply(lambda x: 1 if x == 'Traditional' else 0)
df['Dance'] = df['style'].apply(lambda x: 1 if x == 'Dance' else 0)

df['is_final'] = df['semi_place'].apply(lambda x: 1 if x <= 10 else 0)
```

## Delete NaNs from semi_final column

```python
df.dropna(subset=['semi_place'], inplace=True)
```

## Save new dataset

```python
df.to_csv('preprocessed_data.csv', index=False)
```
